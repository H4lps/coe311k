```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def least_squares(x,y):
    assert(len(x) == len(y))
    n = len(x)
    
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum([x_i**2 for x_i in x])
    sum_xy = sum([x_i * y_i for x_i,y_i in zip(x,y)])
    
    m = (n * sum_xy - (sum_x * sum_y))/((n * sum_x2) - (sum_x)**2)
    c = (sum_y - (m * sum_x))/n
    return m ,c
```

```python
X = np.random.rand(100)
X.sort()
true_m = 5
true_c = 15
pertubations = [np.random.uniform(-1,1) for _ in range(len(X))]
Y =  (true_m * X) + true_c + pertubations

A = np.vstack([X, np.ones(len(X))]).T


m , c = least_squares(X,Y)

x = np.arange(0,1,0.01)

y = m * x + c

print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X,Y)
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Least Squares Fitting")
plt.show()



```

```python
# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

x = np.arange(0,1,0.01)

y = m * x + c

print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X,Y)
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Least Squares Fitting")
plt.show()

```
