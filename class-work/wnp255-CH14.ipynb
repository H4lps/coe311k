```python
#Approximating Derivative of f(x)=sin⁡(x) with Forward Difference, Central Difference, Backwards Difference
import numpy as np
import matplotlib.pyplot as plt


s
f = lambda x : np.sin(x)

# Domain
a = 0
b = 2 * np.pi
N = 50
x = np.linspace(a,b,N)
h = (b - a) / (N - 1)

true_derivative = np.cos(x)

# Central Difference Approximation for the derivative

df_dx_central = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
### YOUR CODE HERE ###

df_dx_forward = (f(x+h) - f(x)) / h
 

# Backward Difference Approximation

### YOUR CODE HERE

df_dx_backward = f(x+h)
 

# Plot the approximation and the true derivative

 

## Do a canvas of 4 subplots or 4 graphs on one plot comparing cos(x), Forward Difference, Backwards Difference, and Central Difference Methods


plt.figure(figsize=(10, 8))

# 1. True derivative
plt.subplot(2, 2, 1)
plt.plot(x, true_derivative)
plt.title("True Derivative: cos(x)")
plt.xlabel("x")
plt.ylabel("y")

# 2. Forward Difference
plt.subplot(2, 2, 2)
plt.plot(x, true_derivative, linestyle='dashed', label="cos(x)")
plt.plot(x, df_dx_forward, label="Forward")
plt.title("Forward Difference")
plt.legend()

# 3. Backward Difference
plt.subplot(2, 2, 3)
plt.plot(x, true_derivative, linestyle='dashed', label="cos(x)")
plt.plot(x, df_dx_backward, label="Backward")
plt.title("Backward Difference")
plt.legend()

# 4. Central Difference
plt.subplot(2, 2, 4)
plt.plot(x, true_derivative, linestyle='dashed', label="cos(x)")
plt.plot(x, df_dx_central, label="Central")
plt.title("Central Difference")
plt.legend()

plt.tight_layout()
plt.show()
```
