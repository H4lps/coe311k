```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5])

Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
x = np.arange(1,5,.1)
y = m * x + c
print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X,Y)
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Least Squares Fitting")
plt.show()
```

```python
X = np.random.randint(0,5,size=5)

Y = np.random.randint(0,5,size=5)

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
x = np.arange(1,5,.1)
y = m * x + c
print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X,Y)
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Least Squares Fitting")
plt.show()
"""
The fitting is singificantly worse with the randomized data in most cases.
This is because majority of the time our data points will not randomly have linear correlation.
"""
```
