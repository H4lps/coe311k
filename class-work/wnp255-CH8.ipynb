```python
import numpy as np
```

```python

def is_diagonal_dominance(A):
    for i in range(A.shape[0]):
        cur = 0
        for j in range(A.shape[1]):
            if (i!=j):
                cur += A[i][j]
        if (cur > abs(A[i][i])):
            return False
    return True
```

```python
A = np.array([[5,1,2],[1,4,2],[1,0,6]])

print(is_diagonal_dominance(A))

B = np.array([[1,1,1],[2,2,2],[3,3,3]])

print(is_diagonal_dominance(B))
```
