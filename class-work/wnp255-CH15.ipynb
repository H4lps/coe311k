```python
import numpy as np
import matplotlib.pyplot as plt


# Define parameters
h_values = np.arange(0.01,1.01,0.05)  # step size
x_start, x_end = 0, 2
baseline_error = None
total_error = 0
def true_solution(x):
    return np.exp(-2 * x)

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y

for h in h_values:
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    
    

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Function representing dy/dx = -2y
    # Implement Euler's method

    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
        
    y_true = true_solution(x_values)
    
    error = np.sum(np.abs(y_true - y_values))
    
    if baseline_error is None:
        baseline_error = error
        
    total_error +=error
    
    print(f"Relative error to baseline is {error - baseline_error} for h= {h}")
    print(f"Current accumulated error is {total_error}")
    
    plt.plot(x_values, y_values, label=f"h={h:.2f}")
    plt.xlabel('x')

    plt.ylabel('y')

    plt.title("Approximation of dy/dx = -2y using Euler's Method")

    plt.legend()

    plt.show()
```
