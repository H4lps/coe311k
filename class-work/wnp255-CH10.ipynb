```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mse(y,y_pred):
    mse = np.mean((y - y_pred)**2)
    return mse
```

```python
def r_squared(y,y_pred):
    ss_total = np.sum((y - np.mean(y))**2)
    ss_res = np.sum((y - y_pred)**2)
    r_squared = 1 - (ss_res / ss_total)
    return r_squared
```

```python
#Part A
#least_squares
X = np.random.uniform(0,5,size=500)

Y = np.random.uniform(0,5,size=500)
# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
y_pred = m * X + c
mse_ls = mse(Y,y_pred)
r_squared_ls = r_squared(Y,y_pred)

#cubic polynomial

poly_coeffs = np.polyfit(X,Y,3)

y_pred = np.polyval(poly_coeffs,X)

mse_p = mse(Y,y_pred)
r_squared_p = r_squared(Y,y_pred)

if r_squared_p > r_squared_ls:
    print("polynomail fit is better in terms of the r-squared quantity")
else:
    print("least squares fit is better in terms of the r-squared quantity")
if mse_p > mse_ls:
    print("polynomial fit is better in terms of the mse quantity")
else:
    print("least squares fit is better in terms of the mse quanity")

```

```python
#Part B

#least_squares
X = np.random.uniform(0,5,size=500)

Y = np.random.uniform(0,5,size=500)
# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
y_pred = m * X + c
mse_ls = mse(Y,y_pred)
r_squared_ls = r_squared(Y,y_pred)

#cubic polynomial
for i in range(3,10):
    poly_coeffs = np.polyfit(X,Y,i)
    y_pred = np.polyval(poly_coeffs,X)
    mse_p = mse(Y,y_pred)
    r_squared_p = r_squared(Y,y_pred)
    print(f"{i} terms")
    print(f"Mean squared error : {mse_p} , R-Squared : {r_squared_p}")
for i in range(3,0,-1):
    poly_coeffs = np.polyfit(X,Y,i)
    y_pred = np.polyval(poly_coeffs,X)
    mse_p = mse(Y,y_pred)
    r_squared_p = r_squared(Y,y_pred)
    print(f"{i} terms")
    print(f"Mean squared error : {mse_p} , R-Squared : {r_squared_p}")
```

```python
#Part C
# Vandermode matrix
X = np.random.uniform(0,5,size=500)

Y = np.random.uniform(0,5,size=500)
# Stack the design matrix with a column of ones (for the intercept)
degree = 3
V = np.vander(X, N=degree+1, increasing=True)

# V.T * V
VT_V = np.dot(V.T, V)
# V.T * y
VT_y = np.dot(V.T, Y)
# Solve for a (coefficients)
coefficients = np.linalg.solve(VT_V, VT_y)

y_pred = np.polyval(coefficients,X)

mse_v = mse(Y,y_pred)

r_squared_v = r_squared(Y,y_pred)

#cubic polynomial

poly_coeffs = np.polyfit(X,Y,3)

y_pred = np.polyval(poly_coeffs,X)

mse_p = mse(Y,y_pred)
r_squared_p = r_squared(Y,y_pred)

if r_squared_p > r_squared_v:
    print("polynomail fit is better in terms of the r-squared quantity")
else:
    print("Vandermonde matrix fit is better in terms of the r-squared quantity")
if mse_p > mse_v:
    print("polynomial fit is better in terms of the mse quantity")
else:
    print("Vandermonde fit is better in terms of the mse quanity")
```
